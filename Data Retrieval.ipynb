```python
prcl = mdb.read_table("prcl.mdb", "PrclREAR")
prcl.head()
```

```python
new_df = pd.DataFrame(prcl, columns=["CityBlock", "Parcel", "OwnerCode", "BillYear","TaxAmt", "TotalValue", "TaxBal", "ResTaxAmt"])
new_df.head()
```

```python
new_df["id"] = new_df["CityBlock"].map(str) + "_" + new_df["Parcel"] + "_" + new_df["OwnerCode"]
```

```python
new_df = pd.DataFrame(new_df, columns=["id", "BillYear","TaxAmt", "TotalValue", "TaxBal", "ResTaxAmt"])
```

```python
filtered_df = new_df.loc[new_df['ResTaxAmt']!= "0.0000"]
filtered_df
```

```python
filtered_df.to_csv('residential_data.csv')
```
